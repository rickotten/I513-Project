## Cryptocurrencies

First here is a short description of what cryptocurrencies are:

A cryptocurrency is a digital currency and asset that is kept track of using a technology known as blockchain. Blockchain is most commonly used as a virtual ledger that keeps track of and records transactions with a cryptocurrency. It is generally considered highly secure as it does not allow counterfeiting and resolves many other issues that arise from the usage of physical currencies. Another appeal of it to many is that cryptocurencies are not associated with any government and therefore can not be manipulated or interfered with by them. 

Cryptocurrencies can be used to purchase the same things you would purchase with any other form of currency. Recently however what people have really been interested in is investing in cryptocurrencies in the same way that people invest in other stocks or goods such as oil and precious metals as it is a very lucrative and expanding market. Many investors would describe the cryptocurrency market as volatile and high-risk and our goal with this dataset is to try and recognize these trends and shifts that causes the cryptocurrency market to be classified in this way.

# Gather all CSVs
Then we created a list of csv file names for automation: 


```python
import os
files = []
for filename in os.listdir(os.getcwd()):
  if filename[-3:] == "csv":
    files += [filename]
print(files)
```

# Descriptions and Initial Statistics of Datasets

Next we used pandas read_csv  command to turn the csv files into a data frame. Then we used pandas .describe()  method to print out summary statistics. We used pandas .datatypes command to print out the data types of the data fields.  


```python
import pandas as pd
import numpy as np

#print out file name, data types, and summary statistics
for f in files:
  df = pd.read_csv(f)
  print(f)
  print('Data Types:')
  print(df.dtypes)
  print('\n')
  print('Statistical Descriptions:')
  print(df.describe())
  print('\n')
```

# Basic Visualizations
We provide the following rough visualizations of basic trends in our dataset. Such trends include Highs vs Lows and Volume vs Market Cap (trends of interest to the consumer). While these are basic, it provides insight into initial patterns and trends we might consider exploring.

```python
# Highs vs Lows over time
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

for f in files:
  df = pd.read_csv(f, parse_dates=["Date"])
  ax = df.plot(x="Date", y=["High", "Low"])
  ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m-%d"))
  ax.xaxis.set_minor_formatter(mdates.DateFormatter("%Y-%m-%d"))
  ax.set_title(f[:-4].split("_")[1])

  
plt.show()
```

```python
# MarketCap vs Volume over time
for f in files:
  df = pd.read_csv(f, parse_dates=["Date"])
  ax = df.plot(x="Date", y=["Marketcap", "Volume"])
  ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m-%d"))
  ax.xaxis.set_minor_formatter(mdates.DateFormatter("%Y-%m-%d"))
  ax.set_title(f[:-4].split("_")[1])

  
plt.show()
```
